### Conditional Statement

```python
## Identify the NUmber Evn or Odd
x = 5
r = x % 2
if r == 0:
    print ("Even Number:",x)
if r == 1:
    print("Odd number:",x)
```

```python
x = 7
r = x % 2
if r == 0:
    print ("Even Number:",x)
else:
    print("Odd number:",x)
```

```python
# Check the Number Even or Odd and Check number is greater than 5
# This 2 conditions

x = 6
r = x % 2
if r == 0:
    print ("Evne Number:",x)
    if x>5:
        print("greater number:",x)
else:
    print("Odd number:",x)
```

```python
# Check the Number Even or Odd and Check number is greater than 5
# This 2 conditions 
# If inside if else is Nested If

x = 9
r = x % 2
if r == 0:
    print ("Evne Number:",x)
    if x>5:
        print("greater number:",x)
    else:
        print("Lesser number:",x)
else:
    print("Odd number:",x)
```

```python
# If Elif Else
x = 2

if x == 1:
    print("one")
elif x == 2:
    print('Two')
elif x == 3:
    print('Three')
else:
    print('Number is not found')
```

### Loops

```python
i = 1 # Initialization
while i<=5: # Condition
    print('data science:', i)
    i = i + 1 # Increment
    
```

```python
i = 5 # Initialization
while i>=1: # Condition
    print('data science:', i)
    i = i - 1 # Decrement
```

```python
# nested while loop
i = 1
while i >= 5:
    print('data science')
    
    i = i +1
    print()
```

```python
i = 1
while i <= 2:
    j = 0
    while j <=2:
        print(i*j, end = " ")
        j +=1
    print()
    i += 1
     
```
