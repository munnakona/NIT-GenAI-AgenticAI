```python
def person(name,age): #positional Arg
    return name
    return age

person('Munna',22)
```

```python
def person(name,age): # keyword Arg
   print(name)
   print (age - 1)

person(name = 'Munna',age = 21)
```

```python
def person(name,age,age2): # keyword Arg
   print(name)
   print (age - 1)
   print (age2 -1)

person(name = 'Munna',age = 21,age2 = 20)
```

```python
def person(name,age=18): #Default Arg
   print(name)
   print (age )

person('Munna',23)
```

```python
def person(name,age): # Variable length  Arg
   print(name)
   print (age - 1)

person(name = 'Munna',age = 21)
```

```python
def add(a, *b):  # Variable length Arg
    c = a
    for i in b:
        c = c+ i
    print(c)

add(22,45,33,24)
```

```python
def person(name,**data): # keyword varaible legth (Kwarg)
    print(name)
    print(data)

person('Munna',age=30, mob=123456)
```
